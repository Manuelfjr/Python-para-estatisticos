**1.5** Escreva um código que encontra todos os divisores do número 42:

```python
def divisor(m = 42):
    a = []
    for i in range(1,m):
        if m % i == 0:
            a.append(i)
    
    a.append(m)
    
    return a

divisor(m = 42)
```

**1.7** Calcule a multiplicação das duas matrizes acima usando laços:

```python
def Prodlac2x2(a,b):
    ab = [[0,0],[0,0]]
    for i in range(0, 2):
        for j in range(0, 2): 
            for k in range(0, 2): # For para realizar as somas , dentro de cada entrada da matriz
                ab[i][j] += a[i][k]*b[k][j]
    return ab 

Prodlac2x2(a,b)
```

**1.8** Calcule o fatorial de 5 usando algum tipo de laço:

```python
def fat(n):
    fator = 1
    for i in range(1, n + 1):
        fator = i * fator
    return fator

fat(5)
```

**1.10** Calcule a distância Euclidiana dos seguintes vetores:

```python
a, b = [1, 0, 1], [-1, 2, 3]

def d(a,b):
    d = 0
    for i in range(0, len(a)):
        d = (a[i]*b[i])**(2) + d
    return (d)**(1/2)

d([1,2,3],[3,2,1])
```

**1.11** Imprima as potências de 2, começando em  20 , enquanto o resultado for menor do que 2000:

```python
def zero(x, min = 20, max = 2000, n = 10):
    for i in range(1,n + 1):
        if x**(i) < min and x**(i) > max:
            break
        else:
            print('{} é equivalente á {} elevado á {}'.format(x**(i), x, i))

zero(2)
```

**1.14** Escreva um laço que resulte no sétimo elemento da sequência de Fibonacci (lembrando que em Python os índices começam em 0):

```python
def fib(n):
    a = 0
    b = 1
    i = 0
    if n == 0:
        return a
    while i + 1 < n:
        a = a + b
        temp = b
        b = a
        a = temp
        i += 1
    
    return a

fib(7)
        
```
